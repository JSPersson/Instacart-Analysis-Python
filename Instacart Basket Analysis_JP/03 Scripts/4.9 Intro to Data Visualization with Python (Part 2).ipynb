# 04. Create Visualizations

### Histograms

```python
# Create histogram for 'prices'column
hist_prices = df_ords_prods_cust['prices'].plot.hist(bins = 100)
```

#### Observations: Right-skewed histogram shows that highest price frequency is less than 15 dollars.

```python
# Create histogram for 'order_hour_of_day'column
hist_order_hour = df_ords_prods_cust['order_hour_of_day'].plot.hist(bins = 70)
```

#### Observations: This left-skewed histogram demonstrates that the highest frequency of sales occurs between 10 and 15 (10:00am-3:00pm), while the lowest frequency of sales occurs between 0 and 6 (12:00am-6:00am). 

### Bar Charts

```python
# Create bar chart for 'orders_day_of_week'
bar_orders_day = df_ords_prods_cust['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

#### Observations: Saturday and Sunday are busiest ordering days.

```python
# Create bar chart for 'loyalty_flag' column
bar_loyalty = df_ords_prods_cust['loyalty_flag'].value_counts().plot.bar()
```

#### Observations: Highest number of customers belong to "Regular Customer" group.

### Sample Subsets for Line Charts

```python
# Create sample dataframes before creating line chart
# Create list with True/False values for test np.random.rant() <=7
np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_cust)) <= 0.7
```

```python
# Store 70% of sample in df_big
df_big = df_ords_prods_cust[dev]
```

```python
# Store 30% of sample in df_small
df_small = df_ords_prods_cust[~dev]
```

```python
# Find number of total records for df_ords_prods_cust
len(df_ords_prods_cust)
```

```python
# Ensure sample dataframes are consistent with total records
len(df_big) + len(df_small)
```

#### Observations: Records for sample dataframes are consistent with total records. Samples successfully created.

### Line Charts

```python
# Create dataframe with only columns necessary for 'prices' across 'orders_day_of_week' line chart
df_4 = df_small[['orders_day_of_week','prices']]
```

```python
# Create line chart for 'orders_day_of_week' and 'prices'
line_hour_prices = sns.lineplot(data = df_4, x = 'orders_day_of_week', y = 'prices')
```

#### Observations:  Most money is spent on 0 (Saturday) and 6 (Friday).

```python
# Create dataframe with only columns necessary for expenditure line chart
df_2 = df_small[['order_hour_of_day','prices']]
```

```python
# Create line chart for 'order_hour_of_day' and 'prices'
line_order_prices = sns.lineplot(data = df_2, x = 'order_hour_of_day', y = 'prices')
```

#### Observations: Expenditure is highest around 3:00am.

```python
# Create dataframe with only columns necessary for family situation line chart
df_3 = df_small[['Age','number_of_dependants']]
```

```python
# Create line chart for 'Age' and "number_of_dependants" using small sample dataframe
line_age_dependants = sns.lineplot(data = df_3, x = 'Age', y = 'number_of_dependants')
```

#### Observations: There doesn't appear to be a correlation between age and the number of dependants. I would expect there to be significantly fewer dependants for customers aged 55+, but I'm not seeing any trends. 

### Scatterplots

```python
# Create scatterplot to identify outliers in 'prices' column
scatter_price = sns.scatterplot(x = 'prices', y = 'prices',data = df_ords_prods_cust)
```

#### Observations: No outliers detected.

```python
# Create scatterplot to visualize relationship between 'Age' and 'income'
scatter_age_income = sns.scatterplot(x = 'Age', y = 'income',data = df_ords_prods_cust)
```

#### Observations: The majority of customers, regardless of age, have an income of less than 200,000. There are many more customers aged 40+ in the 200,000 and higher income range.
